## Question 11: Factorial of a Number
Write a program to find the factorial of a number using a `for` loop.


```python
num = int(input("Enter a number: "))
factorial = 1

if num < 0:
    print("Factorial does not exist for negative numbers")
elif num == 0:
    print("The factorial of 0 is 1")
else:
    for i in range(1, num + 1):
        factorial = factorial * i
    print("The factorial of", num, "is", factorial)
```

## Question 12: Fibonacci Sequence
Write a program to print the Fibonacci sequence up to `n` terms.


```python
n_terms = int(input("How many terms? "))
n1, n2 = 0, 1
count = 0

if n_terms <= 0:
    print("Please enter a positive integer")
elif n_terms == 1:
    print("Fibonacci sequence up to", n_terms, ":")
    print(n1)
else:
    print("Fibonacci sequence:")
    while count < n_terms:
        print(n1, end=" ")
        nth = n1 + n2
        # update values
        n1 = n2
        n2 = nth
        count += 1
```

## Question 13: Prime Number Check
Write a program to check if a number is prime or not.


```python
num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, int(num**0.5) + 1):
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

## Question 14: Reverse a String

## Question 15: Count Vowels
Write a program to count the number of vowels in a string.


```python
string = input("Enter a string: ").lower()
vowels = "aeiou"
count = 0

for char in string:
    if char in vowels:
        count += 1

print("Number of vowels:", count)
```

## Question 16: Largest Element in a List
Write a program to find the largest element in a list.


```python
numbers = [10, 25, 5, 40, 30]
largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("The largest element is:", largest)
```

## Question 17: Sum of Elements in a List
Write a program to find the sum of all elements in a list.


```python
numbers = [1, 2, 3, 4, 5]
total_sum = 0

for num in numbers:
    total_sum += num

print("The sum is:", total_sum)
```

## Question 18: Remove Duplicates from a List
Write a program to remove duplicate elements from a list.


```python
original_list = [1, 2, 2, 3, 4, 4, 5]
unique_list = []

for item in original_list:
    if item not in unique_list:
        unique_list.append(item)

print("Original list:", original_list)
print("List after removing duplicates:", unique_list)
```

## Question 19: Palindrome Check
Write a program to check if a string is a palindrome (reads the same forwards and backwards).


```python
string = input("Enter a string: ").lower()
if string == string[::-1]:
    print(string, "is a palindrome")
else:
    print(string, "is not a palindrome")
```

## Question 20: Student Marks Dictionary
Write a program to create a dictionary of 5 students and their marks, then print the name of the student who scored the highest.


```python
students = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "David": 95,
    "Eve": 88
}

highest_scorer = ""
max_marks = 0

for name, marks in students.items():
    if marks > max_marks:
        max_marks = marks
        highest_scorer = name

print("The student with the highest marks is:", highest_scorer, "with", max_marks, "marks")
```
